# matplotlib.org

```python
# Linear graph 


import matplotlib.pyplot as plt

plt.title('Basic Linear Graph')
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
# Graph to plot distance and time

distance=list(range(2,11,2))
time=list(range(1,6,1))
#d=2*t
plt.title('Distance Vs Time')
plt.xlabel('time')
plt.ylabel('distance')
plt.plot(time,distance,'bo')
plt.grid()
plt.show()
```

```python
# Non linear graph (v=a*t)

t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t))]
plt.plot(t,v)
plt.xlabel('time')
plt.ylabel('velocity')
plt.grid()
plt.show()
```

```python
# Exponential graph
n=list(range(10))
sqn=[i**2 for i in n]
plt.plot(n,sqn)
plt.show()
```

```python
# Sinusoidal wave(Voltage vs Time)

import numpy as np

#t=[i/10 for i in range(1,21)]
t=np.arange(0.1,2.0,0.1)
v=[1+np.sin(2*np.pi*i)  for i in t]
plt.title('SinWave')
plt.plot(t,v)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.savefig('Images/SineWave.png')
plt.show()
```

```python
# Plot Income of Arkanasas from 2005 to 2014
#years=list(range(2005,2014))
import pandas as pd
df=pd.read_csv('DataFiles/Income.csv')
years=df.columns[2:]
incomeArkansas=df.values[3,2:]
plt.title("Income of Arkansas")

plt.xlabel('Years')
plt.ylabel('Income')
plt.plot(years,incomeArkansas)
plt.show()
```

```python
# Income of all states from 2005 to 2013
incomeallstates=[]

for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
plt.plot(years,incomeallstates)
plt.title("Income of all states")
plt.xlabel("Years")
plt.ylabel("Income")
plt.grid()
plt.show()
    
    
```
