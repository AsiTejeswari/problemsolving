```python
# Fucntion to read app store dataset 

import pandas as pd

def readAppleStore(filePath):
    return pd.read_csv(filePath)

filePath = "files/AppleStore.csv"

readAppleStore(filePath)
```

```python
# Function to get all unique prime genre categories

def uniquePrimeGrenres(df):
    return list(df["prime_genre"].unique())
        
        


incomedf = readAppleStore(filePath)

uniquePrimeGrenres(incomedf)
```

```python
# Function to return category with highest number of apps

def highestAppsCategory(df):
    return df["prime_genre"].value_counts().idxmax()+' : '+str(df['prime_genre'].value_counts().max())
                    
incomedf = readAppleStore(filePath)


highestAppsCategory(incomedf)
```

```python
# Function to get category with lowet no of apps

def lowestAppsCategory(df):
    return df["prime_genre"].value_counts().idxmin()+' : '+str(df['prime_genre'].value_counts().min())
    
    
incomedf = readAppleStore(filePath)

lowestAppsCategory(incomedf)
```

```python
# Function to return category with highest user rating

def highestUserRatingCategory(df):
    highRating =  max(list(df["user_rating"]))
    categories = [df['prime_genre'][index] for index in range(len(df["prime_genre"])) if df['user_rating'][index] == highRating]
    return df['prime_genre'].value_counts().idxmax()+' : '+str(df['prime_genre'].value_counts().max())
        

incomedf = readAppleStore(filePath)

highestUserRatingCategory(incomedf)
```

```python
# Apps with highest downloads refers app having highest user rating total

def highestDownloads(df):
    maxRating = max(list(df['rating_count_tot']))
    categories = [df['track_name'][index] for index in range(len(df['rating_count_tot'])) if df['rating_count_tot'][index] == maxRating]
    return categories[0]+' : '+str(maxRating)   
            
        

incomedf = readAppleStore(filePath)

highestDownloads(incomedf)
```

```python
def highestAveargeRatingCount(df):
    uniquePrimeGenres = list(df["prime_genre"].unique())
    ratingFreq = {}
    maxhighestAverageRating = 0
    highestRatingGenre = ""
    for genre in uniquePrimeGenres:
        ratings = [df['rating_count_tot'][index] for index in range(len(df['rating_count_tot'])) if df['prime_genre'][index] == genre]
        avg = sum(ratings)//len(ratings)
        ratingFreq[genre] = avg
        if(avg > maxhighestAverageRating):
            maxhighestAverageRating = avg
            highestRatingGenre = highestRatingGenre.replace(highestRatingGenre, genre)
    return highestRatingGenre+ ' : '+str(maxhighestAverageRating)   

incomedf = readAppleStore(filePath)
    
highestAveargeRatingCount(incomedf)
```

```python
#  Average user rating for free apps

def freeAppsAvgRating(df):
    ratings =  [df['user_rating'][index] for index in range(len(df['user_rating'])) if df['price'][index] == 0]
    return "Free apps avg rating count : "+str(sum(ratings)/len(ratings))

incomedf = readAppleStore(filePath)

freeAppsAvgRating(incomedf)
```

```python
#  Average user rating for Paid apps

def paidAppsAvgRating(df):
    ratings =  [df['user_rating'][index] for index in range(len(df['user_rating'])) if(df['price'][index] > 0)]
    return "Free apps avg rating count : "+str(sum(ratings)/len(ratings))

incomedf = readAppleStore(filePath)

paidAppsAvgRating(incomedf)
```

```python
# Category with highest user avg rating for paid apps

def highestAvgUserRatingPaidCategory(df):
    uniquePrimeGenres = list(df["prime_genre"].unique())
    ratingFreq = {}
    maxhighestAverageUserRating = 0
    highestUserRatingGenre = ""
    for genre in uniquePrimeGenres:
        ratings = [df['user_rating'][index] for index in range(len(df['user_rating'])) if ((df['prime_genre'][index] == genre) and (df['price'][index] > 0))]
        avg = sum(ratings)//len(ratings)
        ratingFreq[genre] = avg
        if(avg > maxhighestAverageUserRating):
            maxhighestAverageUserRating = avg
            highestUserRatingGenre = highestUserRatingGenre.replace(highestUserRatingGenre, genre)
    return highestUserRatingGenre+ ' : '+str(maxhighestAverageUserRating)
    
    

incomedf = readAppleStore(filePath)
    
highestAvgUserRatingPaidCategory(incomedf)
```

```python
# Maximum frequent price point > 0

def maxFrequentPricePoint(df):
    prices = [df['price'][index] for index in range(len(df['price'])) if df['price'][index] > 0]
    maxRepeatedPrice = prices[0]
    repeatedPriceCount = 0
    for price in prices:
        if(prices.count(price)>repeatedPriceCount):
            repeatedPriceCount = prices.count(price)
            maxRepeatedPrice = price
    return str(repeatedPriceCount)

incomedf = readAppleStore(filePath)

maxFrequentPricePoint(incomedf)
```
