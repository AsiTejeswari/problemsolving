### Functions

```python
# function to reverse a string
def reverseString(s):
    return s[::-1]

reverseString("python")
```

```python
# function to check if a string is a palindrom
def palindrom(s):
    if s==s[::-1]:
        return True
    else:
        return False
palindrom("madam")
    
```

```python
#function to check the given year is leapyear or not
def leapyear(year):
    if year%400==0 or (year%100!=0 and year%4==0):
        return True
    else:
        return False
leapyear(2020)
    
```

```python

def digits(n):
    if len(str(n))!=0:
        return len(str(n))
    else:
        return False
digits(12345)
```

```python
#function to identify the greatest of 4 numbers
def largest(n1,n2,n3,n4):
    if n1>n2 and n1>n3 and n1>n4:
        return n1 
    elif n2>n3 and n2>n4:
        return n2 
    elif n3>n4:
        return n3 
    else:
        return n4 
largest(1,2,3,4)
            
```

### Iteration

- for
- while

```python
# function to print n natural no


def naturalno(n):
    for i in range(1,n+1):
        print (i,end=" ")
    return
naturalno(12)
```

```python
# function to print n natural no useing while loop
def naturalno(n):
    i=1
    while i<=n:
        print(i,end = " ")
        i+=1
    return
naturalno(10)
```

```python
# function to print all  numbers  divisible by 6 
# and not a factor of 10 in a given range(lb,ub) inlcusive
def divisible(l,u):
    for i in range(l,u):
        if i%6==0 and i%10!=0:
            print(i)
    return 
divisible(10,100)
```

```python
def avgcubeseven(l,b):
    sum=0
    c=0
    avg=0
    for i in range(l,b+1):
        if i%2==0:
            sum += i**3
            c=c+1
        avg=sum/c
    print(avg)
    return

avgcubeseven(50,75)

    
    
```

```python
# factor list of a given no
def factorlist(n):
    for i in range(1,n+1):
        if n%i==0:
            print(i)
    return
factorlist(12)
```

```python
# function to check if a given no is prime 
def prime(n):
    c=0
    for i in range(1,n+1):
        if n%i==0:
            c+=1
    if c==2:
        print ("it is a prime no")
    else:
        print("not a prime no")
    return
prime(9)
            
```

```python
# function to calculate the average first N prime numbers

def avgofprimes(n):
    c=0
    sum=0
    seqcount=2
    while(c<n):
        if prime(seqcount):
            c+=1
            sum+=seqcount
        seqcount+=1
    return sum/n
avgofprimes(10)
        
        

    
```

```python
# funcion to calculate factoral of a given no
def factorial(n):
    mul=1
    for i in range(1,n+1):
        mul=mul*i
    print(mul)
    return
factorial(5)
```

```python
# function to generate perfect no

def perfectno(n):
    sum=0
    for i in range(1,n):
        if n%i==0:
            sum=sum+i
    print(sum)
    if n==sum:
        print("perfect no")
    else:
        print("not a perfect no")
    return
perfectno(6)
        
            
```

```python
# function to generate all perfect no in the given range
def generateperfect(l,u):
    for i in range(l,u+1):
        if perfectno(i):
            print(i)
    return
generateperfect(1,50)
```
