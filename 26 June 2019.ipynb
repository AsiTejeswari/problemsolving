```python
#cipher
def cipher(s,key):
    text = ""
    for i in s:
        if i.isupper():
            text += chr(65+(ord(i)+key-65) % 26)
        elif i.islower():
            text += chr(97+(ord(i)+key-97) % 26)
        elif i.isnumeric():
            text += str((int(i) + key) % 10)
        else:
            text += i
    print(text)
    return 
s=input()
key=int(input())
cipher(s,key)
```

```python
# Function to finding second largest no in a list

def secondlargest(li):
    # convert the list into unique list
    unique=[]
    for n in li:
        if n not in unique:
            unique.append(n)
    unique=sorted(unique,reverse=True)
    if len(unique)>=2:
        return unique[1]
    else:
        return -1
secondlargest([1,2,4,5,6,8])



    
```

```python
# Function to identify the element with highest frequency in a list
# if many elements have the highest frequency then return highest
#highest frequency of ([     ] )

def highestFreq(li):
    unique={}
    for n in li:
        if  n not in unique:
            unique[n]=1
        else:
            unique[n]+=1
    print(unique)
    # Getting all frequencies into list
    freq=unique.values()
    print(freq)
    maxfreq=max(freq)
    print(maxfreq)
    maxfreqE=[]
    # Identifing the smallest number with maximum 
    for item in unique.items():
        if item[1]==maxfreq:
            maxfreqE.append(item[0])
    # Select the minimum from the keys
    maxfreqE=sorted(maxfreqE) # minfreqE[k-1]
    print(maxfreqE)
    return maxfreqE[4-1]


highestFreq([1,2,3,4,5,1,2,3,4])        
```

```python
# Function to identify second highest element.
# if there are many such elements, return the smallest
# [1,2,3,2,1,4,4,9] --> 
def secondhighestFreq(li):
    unique={}
    for n in li:
        if  n not in unique:
            unique[n]=1
        else:
            unique[n]+=1
    print(unique)
    freq=unique.values()
    print(freq)
    maxfreq=min(freq)
    print(maxfreq)
    maxfreqE=[]
    # Identifing the smallest number with maximum 
    for item in unique.items():
        if item[1]!=maxfreq:
            maxfreqE.append(item[0])
    # Select the minimum from the keys
    maxfreqE=sorted(maxfreqE) # minfreqE[k-1]
    print(maxfreqE)
    return maxfreqE[4-1]

secondhighestFreq([1,2,3,4,5,1,2,3,4,6,7,8,6,1])        
```
